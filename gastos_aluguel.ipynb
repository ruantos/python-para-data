```python
import pandas as pd
```

```python
URL = "https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv"

data = pd.read_csv(URL, sep=";")
```

```python
data.head()
```

### Novas colunas com gastos mensais e anuais por im처vel

```python
data["Valor_por_mes"] = data["Valor"] + data["Condominio"]
```

```python
data["Valor_por_ano"] = (data["Valor_por_mes"] * 12) + data['IPTU']
```

## Colunas categ처ricas

```python
data["Possui_suite"] = data['Suites'] > 0
```

```python
data["Descricao"] = data['Tipo'] + " em " + data['Bairro'] + " com " + str(data['Quartos']) + "quartos" + "e" + str(data['Vagas']) + "vaga(s) de garagem."
```

```python
data.head(5)
```

### Salvando arquivo

```python
data.to_csv("valor_ano_mes.csv", sep=";", index=False)
```
