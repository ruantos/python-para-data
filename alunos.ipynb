### Checando valors NaN

```python
df.isna().sum()
```

```python
df = df.fillna(0)
```

### Removendo alunos que saíram

```python
df.head(18)
```

```python
df = df.query("Nome != 'Alice' & Nome != 'Carlos' ")
```

### Filtrando só os aprovados

```python
aprovados = df.query("Aprovado")
```

### Corrigindo notas 7

```python
df.replace({'Notas': {7: 8}})
```

### Coluna de pontos extras

```python
df["Pontos_extras"] = df["Notas"] * 2 / 5
```

### Coluna notas finais

```python
df["Notas_finais"] = df["Notas"] + df['Pontos_extras']
```

### Coluna aprovação final

```python
df["Aprovado_final"] = df["Notas_finais"] >= 7
```

### Cheque quem passou por recuperação

```python
df.query("~Aprovado & Aprovado_final")
```

### Salvando aprovados em csv

```python
aprovados.to_csv("alunos_aprovados.csv", sep=",", index=False)
```
