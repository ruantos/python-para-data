```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
URL = "https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv"
```

```python
df = np.loadtxt(URL, delimiter=",",
                skiprows=1, usecols=np.arange(1, 6, 1))
```

```python
diametro_lar = df[:5_000, 0]
peso_lar = df[:5_000, 1]

diametro_tor = df[5_000:, 0]
peso_tor = df[5_000:, 1]
```

```python
plt.plot(diametro_lar, peso_lar)
plt.plot(diametro_tor, peso_tor)

plt.xlabel("peso")
plt.ylabel("diâmetro")
plt.title("relação peso por diâmetro")
plt.legend(["laranja", "toranja"])
```
